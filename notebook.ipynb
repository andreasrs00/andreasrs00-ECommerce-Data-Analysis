## Conclusion

1. Faktor utama yang mempengaruhi keterlambatan pengiriman pesanan
Produk dalam kategori tertentu mengalami keterlambatan lebih tinggi dibandingkan yang lain.
Faktor utama keterlambatan meliputi kompleksitas logistik, jarak pengiriman yang jauh, serta volume pesanan tinggi pada periode tertentu.
Estimasi pengiriman yang kurang akurat juga dapat menyebabkan keterlambatan yang konsisten.

2. Pengaruh metode pembayaran terhadap nilai transaksi rata-rata
Metode pembayaran credit_card memiliki nilai transaksi rata-rata tertinggi dibandingkan metode lain seperti boleto atau voucher.
Hal ini bisa dipengaruhi oleh kemudahan pembayaran, batas kredit yang lebih tinggi, serta promo dan opsi cicilan yang tersedia bagi pengguna kartu kredit.
Metode pembayaran alternatif seperti boleto lebih banyak digunakan untuk transaksi dengan nominal lebih kecil.

3. Pelanggan dengan nilai RFM tertinggi dan karakteristik mereka
Sebagian besar pelanggan hanya melakukan satu transaksi dalam periode yang tersedia, seperti yang terlihat dari nilai Frequency dan Monetary yang cenderung 1.
Pelanggan dengan nilai RFM tertinggi adalah mereka yang baru saja melakukan transaksi (recency rendah), sering berbelanja (frequency tinggi), dan memiliki nilai transaksi yang besar (monetary tinggi).
Strategi retensi pelanggan diperlukan untuk meningkatkan loyalitas dan mendorong pembelian berulang.

4. Pola geografis dalam frekuensi dan nilai pesanan
Beberapa kode pos tertentu memiliki volume pesanan yang jauh lebih tinggi dibandingkan wilayah lain, menunjukkan bahwa pesanan terkonsentrasi di area perkotaan atau pusat ekonomi.
Wilayah dengan jumlah pesanan tinggi dapat menjadi target utama untuk optimalisasi logistik dan strategi pemasaran.

5. Pengelompokan kategori produk berdasarkan harga dan volume penjualan
Kategori produk dengan volume penjualan tertinggi cenderung memiliki harga yang lebih rendah, menunjukkan bahwa produk yang lebih terjangkau lebih sering dibeli.
Sebaliknya, beberapa kategori produk dengan harga tinggi memiliki volume penjualan yang lebih rendah tetapi tetap berkontribusi signifikan terhadap total pendapatan.
Hal ini menunjukkan adanya segmen pasar yang berbeda: konsumen yang mencari produk dengan harga terjangkau vs. konsumen yang mengutamakan kualitas dan eksklusivitas.

```python
#Menggabungkan semua data
# Copy dataset agar tidak mengubah aslinya
orders_df_copy = orders_df.copy()

# Tambahkan kolom keterlambatan pengiriman (delivery_delay)
orders_df_copy["delivery_delay"] = (orders_df_copy["order_delivered_customer_date"] - 
                                    orders_df_copy["order_estimated_delivery_date"]).dt.days

# Gabungkan orders dengan customers
all_df = orders_df_copy.merge(customer_df, on="customer_id", how="left")

# Gabungkan dengan order_items untuk mendapatkan detail item pesanan
all_df = all_df.merge(order_item_df, on="order_id", how="left")

# Gabungkan dengan products untuk mendapatkan informasi kategori produk
all_df = all_df.merge(product_df, on="product_id", how="left")

# Gabungkan dengan order_payments untuk mendapatkan informasi pembayaran
all_df = all_df.merge(order_payments_df, on="order_id", how="left")

# Gabungkan dengan sellers untuk mendapatkan informasi penjual
all_df = all_df.merge(sellers_df, on="seller_id", how="left")

# Menghitung RFM (Recency, Frequency, Monetary) berdasarkan hasil analisis sebelumnya
reference_date = orders_df_copy["order_purchase_timestamp"].max()
rfm = all_df.groupby("customer_unique_id").agg(
    Recency=("order_purchase_timestamp", lambda x: (reference_date - x.max()).days),
    Frequency=("order_id", "count"),
    Monetary=("payment_value", "sum")
).reset_index()

# Gabungkan hasil RFM ke dalam all_df
all_df = all_df.merge(rfm, on="customer_unique_id", how="left")

# Menghitung jumlah total pesanan per kategori produk
category_sales = order_item_df.merge(product_df, on="product_id", how="left").groupby("product_category_name").agg(
    avg_price=("price", "mean"),
    total_sales=("product_id", "count")
).reset_index()

# Gabungkan hasil analisis kategori produk ke all_df
all_df = all_df.merge(category_sales, on="product_category_name", how="left")

# Pilih hanya kolom yang relevan untuk dashboard
all_df = all_df[[
    "order_id", "customer_unique_id", "order_status", "order_purchase_timestamp",
    "order_delivered_customer_date", "order_estimated_delivery_date", "delivery_delay",
    "customer_zip_code_prefix", "payment_type", "payment_value",
    "product_id", "product_category_name", "price", "freight_value",
    "seller_id", "seller_zip_code_prefix", "Recency", "Frequency", "Monetary",
    "avg_price", "total_sales"
]]

# Simpan ke CSV
all_df.to_csv("maindata.csv", index=False)
```
